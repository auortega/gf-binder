```gf
abstract Narrative = {
    flags startcat = Narrative;

    cat
    Narrative ;
    Intro ; 
    Analysis_Phrase ;
    Element ;
    Trend_Phrase ; 
    Observation ; 
    Timeframe ; 
    Trend ; 
    Change ; 
    Analysis ; 
    Value_Obs ; 
    Value ; 
    Unit ; 
    Year ; 
    Year1 ;
    Year2 ;
    Str_shows ;
    Str_with ;
    Str_between ;
    Str_theYears ;
    Str_and ;
    Str_of ;
    Str_for ;
    Str_forYear ;

    fun
    describe : Intro -> Analysis_Phrase -> Narrative ;
    shows: Element -> Str_shows -> Intro ;
    analize : Trend_Phrase -> Str_with -> Observation -> Timeframe -> Analysis_Phrase ;
    analize2 : Trend_Phrase -> Str_with -> Observation -> Analysis_Phrase ;
    trends : Trend -> Change -> Trend_Phrase ;
    observed : Analysis -> Str_of -> Value_Obs -> Observation ;
    with_value : Value -> Unit -> Value_Obs ;
    between : Str_between -> Str_theYears -> Year1 -> Str_and -> Year2 -> Timeframe ;
    for_year1 : Str_for -> Year -> Timeframe ;
    for_year2 : Str_forYear -> Year -> Timeframe ;
    
    element : Element ;
    trend : Trend ;
    change : Change ;
    analysis : Analysis ;
    value : Value ;
    unit : Unit ;
    year : Year ;
    year1 : Year1 ;
    year2 : Year2 ;
    
    l_shows : Str_shows ;
    l_with : Str_with ;
    l_between : Str_between ;
    l_theYears : Str_theYears ;
    l_and : Str_and ;
    l_of: Str_of ;
    l_for : Str_for ;
    l_forYear : Str_forYear ;
    
}
```
